# LAB | Imbalanced

**Load the data**

In this challenge, we will be working with Credit Card Fraud dataset.

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv

Metadata

- **distance_from_home:** the distance from home where the transaction happened.
- **distance_from_last_transaction:** the distance from last transaction happened.
- **ratio_to_median_purchase_price:** Ratio of purchased price transaction to median purchase price.
- **repeat_retailer:** Is the transaction happened from same retailer.
- **used_chip:** Is the transaction through chip (credit card).
- **used_pin_number:** Is the transaction happened by using PIN number.
- **online_order:** Is the transaction an online order.
- **fraud:** Is the transaction fraudulent. **0=legit** -  **1=fraud**


```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.utils import resample
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fraud = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/card_transdata.csv")
fraud.head()
```

**Steps:**

- **1.** What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?
- **2.** Train a LogisticRegression.
- **3.** Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.
- **4.** Run **Oversample** in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model? 
- **5.** Now, run **Undersample** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?
- **6.** Finally, run **SMOTE** in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model? 

```python
#1. What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset -> YES theres much less fraud data
sns.countplot(x='fraud', data=fraud, palette=['#4CAF50', '#F44336']) 
plt.title('Fraud vs Legit Distribution')
plt.show()
```

```python
#2. Train a LogisticRegression.
features = fraud.drop('fraud', axis=1)
target = fraud['fraud']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42)

lr = LogisticRegression ()
lr.fit(X_train, y_train)
```

```python
# 3. Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.
lr_pred = lr.predict(X_test)
print("--- Logistic Regression Results ---")
print(f"Accuracy: {accuracy_score(y_test, lr_pred):.2%}")
print(classification_report(y_test, lr_pred))
#the models recall for Macro Average (fraud detection) is 79% and low. We can improve this by over/undersampling. 
```

```python
# 4. Run Oversample in order to balance our target variable and repeat the steps above, now with balanced data. 
#first, scale 
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled_np = scaler.transform(X_train)
X_test_scaled_np = scaler.transform(X_test)

#creating a new dataframe for scaled values 
train = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
#creating the column 
train["fraud"] = y_train.values

fraud = train[train["fraud"] == 1]
not_fraud = train[train["fraud"] == 0]

#executing oversamling 
yes_oversampled = resample(fraud,
                                    replace=True,
                                    n_samples = len(not_fraud),
                                    random_state=0)
train_over = pd.concat([yes_oversampled, not_fraud])
train_over

#check if oversamling worked visually 
fraud_plt = train_over["fraud"].value_counts()
fraud_plt.plot(kind="bar")
plt.show()

#Does it improve the performance of our model? => YES the recall macro avg increased from 79% to 94% 
#new oversampled featuers & target 
X_train_ovs = train_over.drop('fraud', axis=1) 
y_train_ovs = train_over['fraud']              

#fit the model with oversampled data 
lr_oversampled = LogisticRegression()
lr_oversampled.fit(X_train_ovs, y_train_ovs)

#separate the fearures and target again
X_train_bal = train_over.drop('fraud', axis=1)
y_train_bal = train_over['fraud']

lr_balanced = LogisticRegression(max_iter=1000)
lr_balanced.fit(X_train_bal, y_train_bal)

#scale again
X_test_scaled = scaler.transform(X_test)

#generate the new report 
y_pred = lr_balanced.predict(X_test_scaled)

print("--- Final Oversampled Model Report ---")
print(classification_report(y_test, y_pred))
```
